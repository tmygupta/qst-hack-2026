```python
import numpy as np
from mrmustard.lab.states import Coherent, Number, SqueezedVacuum, Vacuum
from mrmustard.lab.transformations import BSgate, Sgate, Dgate


# Create cat states
cat_horizontal = (Coherent(mode=0, alpha=2) + Coherent(mode=0, alpha=-2)).normalize()
cat_vertical = (Coherent(mode=1, alpha=2j) + Coherent(mode=1, alpha=-2j)).normalize()

# merge with beamsplitter
both_modes = cat_vertical >> cat_horizontal >> BSgate(modes=(0, 1), theta=np.pi/4)

# Wigner function of the marginal
both_modes[0]
```
